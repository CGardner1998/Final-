```python
import pandas as pd

file_path = r'C:\Users\cheyanne.gardner\OneDrive\Documents\DeVos\Restaurant_Reviews.tsv'
data = pd.read_csv(file_path, sep='\t')


data.head()
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import CountVectorizer
import seaborn as sns
import pandas as pd

def generate_wordcloud(text):
    wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)
    plt.figure(figsize=(10, 6))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis("off")
    plt.title("Word Cloud of Restaurant Reviews")
    plt.show()

def plot_top_bigrams(reviews, n=10):
    vectorizer = CountVectorizer(ngram_range=(2, 2), stop_words='english')
    bigram_matrix = vectorizer.fit_transform(reviews)
    bigram_freq = pd.DataFrame(bigram_matrix.sum(axis=0), columns=vectorizer.get_feature_names_out()).T
    bigram_freq.columns = ['Frequency']
    bigram_freq = bigram_freq.sort_values('Frequency', ascending=False).head(n)
    
    plt.figure(figsize=(10, 6))
    sns.barplot(x=bigram_freq.Frequency, y=bigram_freq.index, orient="h")
    plt.title("Top Bigrams in Restaurant Reviews")
    plt.xlabel("Frequency")
    plt.ylabel("Bigram")
    plt.show()

def plot_sentiment_distribution(data):
    plt.figure(figsize=(6, 4))
    sns.countplot(data=data, x="Liked", palette="viridis")
    plt.title("Sentiment Distribution in Reviews")
    plt.xticks(ticks=[0, 1], labels=["Negative", "Positive"])
    plt.xlabel("Sentiment")
    plt.ylabel("Count")
    plt.show()

text = " ".join(data['Review'])
generate_wordcloud(text)
plot_top_bigrams(data['Review'])
plot_sentiment_distribution(data)

```
